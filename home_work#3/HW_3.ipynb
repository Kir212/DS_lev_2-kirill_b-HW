```python
import cv2
import numpy as np
import base64
import torch
from ultralytics import YOLO
from IPython.display import display, Javascript, HTML

# 1. Загружаем модель со свежими весами из папки train-2
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = YOLO('/content/runs/detect/train/weights/best.pt').to(device)

# ГАРАНТИРОВАННОЕ ИСПРАВЛЕНИЕ: Перезаписываем имя во всех структурах YOLO
model.names[0] = "Ricky"
if hasattr(model, 'model') and hasattr(model.model, 'names'):
    model.model.names[0] = "Ricky"


# 2. Создаем HTML-тег для вывода видео
display(HTML('<h3>Рики-Локатор v2.0 (Успешное обучение):</h3><img id="yolo_output" style="border: 2px solid #00ff00; width: 100%; max-width: 640px; height: auto;"/>'))

# 3. JS-функция для трансляции веб-камеры
def video_stream():
    js = Javascript('''
    var video;
    var div = document.createElement('div');
    var stream;

    async function init() {
        video = document.createElement('video');
        video.style.display = 'block';
        video.style.width = '320px'; // маленькое превью оригинального потока

        try {
            stream = await navigator.mediaDevices.getUserMedia({video: true});
        } catch(err) {
            alert("Ошибка доступа к камере! Дайте разрешение в браузере: " + err);
            return;
        }

        document.body.appendChild(div);
        div.appendChild(video);
        video.srcObject = stream;
        await video.play();

        var canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;

        var btn = document.createElement('button');
        btn.textContent = 'Остановить камеру';
        div.appendChild(btn);

        var stop = false;
        btn.onclick = () => { stop = true; };

        var imgOutElement = document.getElementById("yolo_output");

        while (!stop) {
            canvas.getContext('2d').drawImage(video, 0, 0);
            var imgData = canvas.toDataURL('image/jpeg', 0.7);

            var result = await google.colab.kernel.invokeFunction('notebook.run_yolo', [imgData], {});

            if (result && result.data && result.data['text/plain']) {
                var raw_str = result.data['text/plain'].replace(/['"]/g, '');
                if (imgOutElement) {
                    imgOutElement.src = "data:image/jpeg;base64," + raw_str;
                }
            }
            await new Promise(resolve => setTimeout(resolve, 30));
        }

        stream.getTracks().forEach(track => track.stop());
        div.remove();
    }
    init();
    ''')
    display(js)

# 4. Высокоскоростная функция обработки кадра (conf=0.10)
def run_yolo(img_base64):
    try:
        if isinstance(img_base64, list):
            img_base64 = img_base64.pop(0)

        img_bytes = base64.b64decode(img_base64.split(',').pop(1))
        np_arr = np.frombuffer(img_bytes, np.uint8)
        frame = cv2.imdecode(np_arr, cv2.IMREAD_COLOR)

        results = model(frame, conf=0.08, imgsz=320, verbose=False)

        single_result = results.pop(0)
        annotated_frame = single_result.plot()

        _, encoded_img = cv2.imencode('.jpg', annotated_frame)
        img_out = base64.b64encode(encoded_img).decode('utf-8')

        return img_out

    except Exception as e:
        print(f"Ошибка обработки: {e}")
        return ""

# Регистрируем функцию в системе Colab и запускаем трансляцию
from google.colab import output
output.register_callback('notebook.run_yolo', run_yolo)
video_stream()
```
